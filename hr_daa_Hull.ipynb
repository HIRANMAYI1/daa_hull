```python
import matplotlib.pyplot as plt
import random
import time

# Define a Point class to represent 2D points
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

# Function to calculate the orientation of three points (p, q, r)
# Returns 0 if they are collinear, 1 if clockwise, and -1 if counterclockwise.
def orientation(p, q, r):
    val = (q.y - p.y) * (r.x - q.x) - (q.x - p.x) * (r.y - q.y)
    if val == 0:
        return 0
    return 1 if val > 0 else -1

# Function to compute the convex hull of a set of points using the Graham's Scan algorithm
def convex_hull(points):
    # Sort points by x-coordinate to find the leftmost point
    points.sort(key=lambda p: (p.x, p.y))
    n = len(points)

    if n <= 3:
        return points

    hull = []

    # Build the upper hull
    def build_upper_hull():
        for p in points:
            # Check if the last two points in the hull and the current point form a left turn
            while len(hull) >= 2 and orientation(hull[-2], hull[-1], p) != -1:
                hull.pop()  # Remove the last point from the hull if not part of the convex hull
            hull.append(p)

    build_upper_hull()

    # Reverse the points to build the lower hull
    points.reverse()
    hull.pop()  # Remove the duplicated point at the end of the upper hull

    # Build the lower hull using the same logic as the upper hull
    build_upper_hull()

    return hull

# Generate random points
random.seed(42)
num_points = 50
points = [Point(random.uniform(0, 100), random.uniform(0, 100)) for _ in range(num_points)]

# Start measuring time
start_time = time.time()

# Compute the convex hull
convex_hull_points = convex_hull(points)

# End measuring time
end_time = time.time()

# Calculate the elapsed time
elapsed_time = end_time - start_time
print(f"Time taken to compute the convex hull: {elapsed_time:.6f} seconds")

# Extract x and y coordinates for plotting
x = [point.x for point in convex_hull_points]
y = [point.y for point in convex_hull_points]

# Plot the convex hull
plt.figure(figsize=(8, 8))
plt.plot(x, y, marker='o', linestyle='-', color='b', markersize=6, label='Convex Hull')

# Plot the original points (optional)
original_x = [point.x for point in points]
original_y = [point.y for point in points]
plt.scatter(original_x, original_y, color='r', label='Original Points', s=20)

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Convex Hull of Random Points')
plt.legend()
plt.grid(True)
plt.show()
```
